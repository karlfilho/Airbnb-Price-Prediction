```python
import pandas as pd
```

```python
# Load DataFrame 'dataset' and drop first column
dataset = pd.read_csv('final_df.csv', encoding='utf-8', sep=';')
dataset.drop(dataset.columns[0], axis=1, inplace=True)

dataset.head(3)
```

```python
dataset.info()
```

```python
# Droping more columns...
dataset.drop(dataset.columns[5], axis=1, inplace= True)
dataset.info()
```

```python
# Finding registers with null values
nulls = dataset.isnull().sum()
nulls[nulls > 0]
```

```python
# Treating null values with mean (for numerical columns) and with mode (for categorical columns))
for column in dataset.columns:
    if dataset[column].dtype in ['float64', 'int64']:
        # Treating with mean for numerical columns
        dataset[column].fillna(dataset[column].mean())
    else:
        # Treating with mode for categorical columns
        dataset[column].fillna(dataset[column].mode()[0])

# Checking for reminiscent null values
nulls_ = dataset.isnull().sum()
nulls_[nulls_ > 0]
```

```python
# There are some registries that hahaves a price of just R$ 1.00, which doesn't make sense.
# Let's do a sanity check before going forward

wrong_price_registries = dataset[dataset['roomPrice'] <=80]
dataset.drop(wrong_price_registries.index, inplace=True)

print(dataset[dataset['roomPrice'] <= 80])
```

```python
# Descriptive statistics for numerical columns
num_dataset = dataset[['roomPrice', 'rating', 'numberReviews']]
desc_stats = num_dataset.describe()

desc_stats
```

```python
dataset.to_csv('cleaned_final_df.csv', sep=';')
```
